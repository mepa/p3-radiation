```python
%matplotlib inline
import yt
import numpy as np
import matplotlib.pyplot as plt
from yt import derived_field
from yt.units import g, pc, kpc
from yt.utilities.physical_constants import mp, me, Na
```

```python
print 1 / Na
print mp
```

```python
e_abar = me / mp
Ah = 1.0 + e_abar
Ahplu = 1.0
Ahmin = 1.0 + 2.0 * e_abar
Ahel = 4.0 + 2.0 * e_abar
Ahep = 4.0 + 1.0 * e_abar
Ahepp = 4.0
Ahtwo = 2.0 + 2.0 * e_abar
Ahtwp = 2.0 + e_abar
Adeut = 2.0 + e_abar
Adplu = 2.0
Ahd = 3.0 + 2.0 * e_abar
Aelec = e_abar
```

```python
@derived_field(name = "number_density", units = "1/cm**3", force_override=True)
def _numdens(field, data):
    return data["density"] / mp
@derived_field(name = "number_density_h", units = "1/cm**3", force_override=True)
def _numdensh(field, data):
    return data["density"] * data["h   "] / (Ah * mp)
@derived_field(name = "number_density_hplu", units = "1/cm**3", force_override=True)
def _numdenshplu(field, data):
    return data["density"] * data["hplu"] / (Ahplu * mp)
@derived_field(name = "number_density_hel", units = "1/cm**3", force_override=True)
def _numdenshel(field, data):
    return data["density"] * data["hel "] / (Ahel * mp)
@derived_field(name = "number_density_hep", units = "1/cm**3", force_override=True)
def _numdenshep(field, data):
    return data["density"] * data["hep "] / (Ahep * mp)
@derived_field(name = "number_density_hepp", units = "1/cm**3", force_override=True)
def _numdenshepp(field, data):
    return data["density"] * data["hepp"] / (Ahepp * mp)
@derived_field(name = "number_density_htwo", units = "1/cm**3", force_override=True)
def _numdenshtwo(field, data):
    return data["density"] * data["htwo"] / (Ahtwo * mp)
@derived_field(name = "number_density_htwp", units = "1/cm**3", force_override=True)
def _numdenshtwp(field, data):
    return data["density"] * data["htwp"] / (Ahtwp * mp)
@derived_field(name = "number_density_deut", units = "1/cm**3", force_override=True)
def _numdensdeut(field, data):
    return data["density"] * data["deut"] / (Adeut * mp)
@derived_field(name = "number_density_dplu", units = "1/cm**3", force_override=True)
def _numdensdplu(field, data):
    return data["density"] * data["dplu"] / (Adplu * mp)
@derived_field(name = "number_density_hd", units = "1/cm**3", force_override=True)
def _numdenshd(field, data):
    return data["density"] * data["hd  "] / (Ahd * mp)
@derived_field(name = "number_density_hmin", units = "1/cm**3", force_override=True)
def _numdenshmin(field, data):
    return data["density"] * data["hmin"] / (Ahmin * mp)
@derived_field(name = "number_density_elec", units = "1/cm**3", force_override=True)
def _numdenselec(field, data):
    return data["density"] * data["elec"] / (Aelec * mp)
```

```python
@derived_field(name = "number_density_h_nuclei", units = "1/cm**3", force_override=True)
def _numdenshnuc(field, data):
    return data["number_density_h"] + data["number_density_hplu"] + 2.0 * data["number_density_htwo"]
```

```python
@derived_field(name = "abundance_h", units = "1", force_override=True)
def _abundh(field, data):
    return data["h   "] / Ah
@derived_field(name = "abundance_hplu", units = "1", force_override=True)
def _abundhplu(field, data):
    return data["hplu"] / Ahplu
@derived_field(name = "abundance_hel", units = "1", force_override=True)
def _abundhel(field, data):
    return data["hel "] / Ahel
@derived_field(name = "abundance_hep", units = "1", force_override=True)
def _abundhep(field, data):
    return data["hep "] / Ahep
@derived_field(name = "abundance_hepp", units = "1", force_override=True)
def _abundhepp(field, data):
    return data["hepp"] / Ahepp
@derived_field(name = "abundance_htwo", units = "1", force_override=True)
def _abundhtwo(field, data):
    return data["number_density_htwo"] / data["number_density_h_nuclei"]
@derived_field(name = "abundance_htwo_test", units = "1", force_override=True)
def _abundhtwotest(field, data):
    return data["htwo"] / Ahtwo
@derived_field(name = "abundance_htwp", units = "1", force_override=True)
def _abundhtwp(field, data):
    return data["htwp"] / Ahtwp
@derived_field(name = "abundance_deut", units = "1", force_override=True)
def _abunddeut(field, data):
    return data["deut"] / Adeut
@derived_field(name = "abundance_dplu", units = "1", force_override=True)
def _abunddplu(field, data):
    return data["dplu"] / Adplu
@derived_field(name = "abundance_hd", units = "1", force_override=True)
def _abundhd(field, data):
    return data["hd  "] / Ahd
@derived_field(name = "abundance_hmin", units = "1", force_override=True)
def _abundhmin(field, data):
    return data["hmin"] / Ahmin
@derived_field(name = "abundance_elec", units = "1", force_override=True)
def _abundelec(field, data):
    return data["elec"] / Aelec
```

# Radius vs. Ionizing Species Abundances

```python
rmin = 1.0
rmax = 1.0e3
runit = 'pc'
```

```python

prof_h = yt.create_profile(sp, 'radius', ('flash', 'h   '),
                          units = {'radius': runit},
                          extrema = {'radius': ((rmin,runit),(rmax,runit))})
prof_hel = yt.create_profile(sp, 'radius', ('flash', 'hel '),
                          units = {'radius': runit},
                          extrema = {'radius': ((rmin,runit),(rmax,runit))})
prof_hplu = yt.create_profile(sp, 'radius', ('flash', 'hplu'),
                          units = {'radius': runit},
                          extrema = {'radius': ((rmin,runit),(rmax,runit))})
prof_hep = yt.create_profile(sp, 'radius', ('flash', 'hep '),
                          units = {'radius': runit},
                          extrema = {'radius': ((rmin,runit),(rmax,runit))})                                                              
prof_hepp = yt.create_profile(sp, 'radius', ('flash', 'hepp'),
                          units = {'radius': runit},
                          extrema = {'radius': ((rmin,runit),(rmax,runit))})                                                             
                                                                
```

```python
radius = prof_h.x
h = prof_h['flash','h   '].value
hel = prof_hel['flash','hel '].value
hplu = prof_hplu['flash','hplu'].value
hep = prof_hep['flash','hep '].value
hepp = prof_hepp['flash','hepp'].value
```

```python
plt.loglog(radius, h, label="H")
plt.loglog(radius, hel, label="He")
plt.loglog(radius, hplu, label="H$^+$")
plt.loglog(radius, hep, label="He$^+$")
plt.loglog(radius, hepp, label="He$^{++}$")
plt.xlabel("Radius (pc)")
plt.ylabel("Mass Fractions")
plt.legend()
plt.savefig(outdir + 'profile_abundance_ioniz_' + prefix + str(chk) + '_z' + redshift + '.png')
```
